Step 3: Data Cleaning and Transformation


```python
df['Price'].fillna(df['Price'].median(), inplace=True)
df['Developers'].fillna('Unknown', inplace=True)
df
```

```python
df['Release date'] = pd.to_datetime(df['Release date'], format='%b %d, %Y', errors='coerce')
df['Estimated owners'].head()
```

```python
# Step 2: Handle Estimated owners
valid_mask = df['Estimated owners'].str.contains(' - ', na=False)
df_valid = df[valid_mask]

# Fill missing or inconsistent values
df['Estimated owners'] = df['Estimated owners'].fillna("0 - 0")

# Split Estimated owners
df[['Min owners', 'Max owners']] = (
    df['Estimated owners']
    .str.split(' - ', expand=True)
    .astype(float)
)
```

```python
df['Genres'] = df['Genres'].str.lower()
```

```python
df['Tags'] = df['Tags'].str.lower()
```
